```python
# Import all libraries and dependencies
import pandas as pd
import os
import numpy as np
import scipy as scipy
import matplotlib.pyplot as plt
import matplotlib.figure as figure
```

```python
# Import CSV files
education_environment_df = pd.read_csv("../Resources/EducationAndEnviron_Data.csv", delimiter=';').dropna()
public_sector_df = pd.read_csv("../Resources/PublicSector_Indicators.csv", delimiter=';')
```

```python
education_environment_df.head()
```

```python
carbon_emissions = education_environment_df.loc[(education_environment_df['SeriesCode'] == 'EN.ATM.CO2E.PC') & (education_environment_df['Year'] >= 2010)]

carbon_emissions.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('CO2 emissions (metric tons per capita)')
plt.legend(title="Country", loc="upper right")

plt.show()
```

```python
rural_population = education_environment_df.loc[(education_environment_df['SeriesCode'] == 'SP.RUR.TOTL.ZS') & (education_environment_df['Year'] >= 2010)]

rural_population.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Rural population (% of total population)')
plt.legend(title="Country", loc="upper right")

plt.show()
```

```python
urban_population = education_environment_df.loc[(education_environment_df['SeriesCode'] == 'SP.URB.TOTL.IN.ZS') & (education_environment_df['Year'] >= 2010)]

urban_population.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Urban population (% of total population)')
plt.legend(title="Country", loc="upper right")

plt.show()
```

```python
electricty_production_renewable = education_environment_df.loc[(education_environment_df['SeriesCode'] == 'EG.ELC.RNWX.ZS') & (education_environment_df['Year'] >= 2010)]

electricty_production_renewable.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Electricity production from renewable sources (% of total)')
plt.legend(title="Country", loc="upper right")

plt.tight_layout()

plt.show()
```

```python
employment_agriculture = public_sector_df.loc[(public_sector_df['SeriesCode'] == 'SL.AGR.EMPL.ZS') & (public_sector_df['Year'] >= 2010)]

employment_agriculture.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Employment in agriculture (% of total employment)')
plt.legend(title="Country", loc="upper right")

plt.show()
```

```python
employment_industry = public_sector_df.loc[(public_sector_df['SeriesCode'] == 'SL.IND.EMPL.ZS') & (public_sector_df['Year'] >= 2010)]

employment_industry.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Employment in industry (% of total employment)')
plt.legend(title="Country", loc="upper right")

plt.show()
```

```python
employment_services = public_sector_df.loc[(public_sector_df['SeriesCode'] == 'SL.SRV.EMPL.ZS') & (public_sector_df['Year'] >= 2010)]

employment_services.pivot(index='Year', columns='CountryName', values='Value').plot(kind='line')

plt.ylabel('Employment in services (% of total employment)')
plt.legend(title="Country", loc="upper right")

plt.show()
```
